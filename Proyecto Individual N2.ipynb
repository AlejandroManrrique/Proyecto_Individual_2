Vamos a calcular el primer KPI, el cual consta de Reducir en un 10% la tasa de homicidios en siniestros viales de los últimos seis meses, en CABA, en comparación con la 
tasa de homicidios en siniestros viales del semestre anterior.

Definimos a la tasa de homicidios en siniestros viales como el número de víctimas fatales en accidentes de tránsito por cada 100,000 habitantes en un área geográfica durante
un período de tiempo específico. Su fórmula es: (Número de homicidios en siniestros viales / Población total) * 100,000. Tomamos como 3000000 de habitantes la poblacion de CABA

```python
import pandas as pd



# Creamos una columna 'Semestre' en base al MES
kpi1_df['Semestre'] = (df_hechos_victimas['MES'] - 1) // 6 + 1

# Agrupamos por año y semestre y calculamos estadísticas.
kpi1_df = df_hechos_victimas.groupby(['AÑO', 'Semestre']).agg(
    N_VICTIMAS=('N_VICTIMAS', 'sum'),
).reset_index()

# Definimos la poblácion total
poblacion_total = 3000000

# Calculamos la tasa de homicidios en siniestros viales para cada semestre
kpi1_df['Tasa_Homicidios'] = (kpi1_df['N_VICTIMAS'] / poblacion_total) * 100000

# Calculamos el cambio porcentual
kpi1_df['Cambio_Porcentual'] = kpi1_df['Tasa_Homicidios'].pct_change() * 100



kpi1_df
```

La condición requerida se cumple en el 1er semestre del 2017, 1er y 2do semestre del 2019, en el 1er semestre del 2020 y en el ultimo semestre del 2021 

Observamos que entre el primer semestre del 2016 y el 2do semestre del 2018 hay osiclaciones en el numero de victimas, luego aparece una marcada tendencia a la baja, hasta lo que
seria el principio de la pandemia en el primer semestre del 2020. Si bien luego se produce una alza porcentual, los valores mantienen la tendencia de baja que tenian antes
de la pandemia. 

Vamos a calcular el segundo KPI el cual consiste en Reducir en un 7% la cantidad de accidentes mortales de motociclistas en el último año, en CABA, respecto al año anterior

Definimos a la cantidad de accidentes mortales de motociclistas en siniestros viales como el número absoluto de accidentes fatales en los que estuvieron involucradas 
víctimas que viajaban en moto en un determinado periodo temporal. Su fórmula para medir la evolución de los accidentes mortales con víctimas en moto es: 
(Número de accidentes mortales con víctimas en moto en el año anterior - Número de accidentes mortales con víctimas en moto en el año actual) / (Número de accidentes mortales con víctimas en moto en el año anterior) * 100

```python


# Creamos un DataFrame para almacenar los resultados del KPI2
kpi2_df = pd.DataFrame(columns=['Año', 'Accidentes_Mortales_Moto', 'Cambio', 'Cambio_Porcentual'])

# Definimos los años
años = [2016, 2017, 2018, 2019, 2020, 2021]

# Calcular los KPIs para cada año
for i in range(len(años)):
    año_actual = años[i]

    # Filtramos datos para el año actual
    df_año_actual = df_hechos_victimas[(df_hechos_victimas['AÑO'] == año_actual) & (df_hechos_victimas['VICTIMA'] == 'MOTO')]

    # Contamos la cantidad de accidentes mortales de motociclistas para el año actual
    accidentes_mortales_año_actual = df_año_actual.shape[0]

    # Verificamos si estamos en el año 2016
    if año_actual == 2016:
        cambio = None
        cambio_porcentual = None
    else:
        # Filtramos datos para el año anterior
        año_anterior = años[i - 1]
        df_año_anterior = df_hechos_victimas[(df_hechos_victimas['AÑO'] == año_anterior) & (df_hechos_victimas['VICTIMA'] == 'MOTO')]
        
        # Contamos la cantidad de accidentes mortales de motociclistas para el año anterior
        accidentes_mortales_año_anterior = df_año_anterior.shape[0]
        
        # Calculamos el cambio y el cambio porcentual
        cambio = accidentes_mortales_año_anterior - accidentes_mortales_año_actual
        cambio_porcentual = ((accidentes_mortales_año_actual - accidentes_mortales_año_anterior) / accidentes_mortales_año_anterior) * 100

    # Agregamos los resultados al DataFrame
    kpi2_df = pd.concat([kpi2_df, pd.DataFrame({'Año': [año_actual], 'Accidentes_Mortales_Moto': [accidentes_mortales_año_actual], 'Cambio': [cambio], 'Cambio_Porcentual': [cambio_porcentual]})], ignore_index=True)

# Imprimir el DataFrame KPI2
kpi2_df


```

Vemos que el 2017, 2019 y 2020 cumplen con el requisito, las bajas en porcentaje fueron mas importantes que las subas hasta el 2019, de todas formas el año de pandemia modifico
bruscamente los resultados primero a la baja porcentual y luego hacia una alza importante

Para el tercer KPI vamos a ver que comuna tiene la mayor cantidad de victimas, y si redujo su porcentaje a través de los años.

```python


# Agrupamos el DataFrame por la columna "COMUNA" y sumamos el número de víctimas para cada comuna
comuna_victimas = df_hechos_victimas.groupby('COMUNA')['N_VICTIMAS'].sum().reset_index()

# Ordenamos el DataFrame resultante en orden descendente
comuna_victimas_ordenadas = comuna_victimas.sort_values(by='N_VICTIMAS', ascending=False)

# Imprimimos el DataFrame con todas las comunas ordenadas por número de víctimas
print(comuna_victimas_ordenadas)
```

Observamos que la comuna 1 es la que mas víctimas tiene, veremos como fue su comportamiento con los años, si efectivamente redujo la cantidad

```python


# Crear un DataFrame para almacenar los resultados del KPI
kpi3_comuna1_df = pd.DataFrame(columns=['Año', 'Accidentes_Mortales', 'Cambio', 'Cambio_Porcentual'])

# Definir los años de interés
años = [2016, 2017, 2018, 2019, 2020, 2021]

# Calcular los KPIs para cada año
for año_actual in años:
    año_anterior = año_actual - 1

    # Filtrar datos para la comuna 1 y el año anterior y actual
    df_año_anterior = df_hechos_victimas[(df_hechos_victimas['COMUNA'] == 1) & (df_hechos_victimas['AÑO'] == año_anterior)]
    df_año_actual = df_hechos_victimas[(df_hechos_victimas['COMUNA'] == 1) & (df_hechos_victimas['AÑO'] == año_actual)]

    # Contar la cantidad de accidentes mortales de todas las víctimas para cada año
    accidentes_mortales_año_anterior = df_año_anterior.shape[0]
    accidentes_mortales_año_actual = df_año_actual.shape[0]

    # Verificar si estamos en el año 2016
    if año_actual == 2016:
        cambio_porcentual = None  # No calculamos el cambio en 2016
        cambio = None
    else:
        # Calcular el cambio y el cambio porcentual
        cambio = accidentes_mortales_año_actual - accidentes_mortales_año_anterior
        cambio_porcentual = (cambio / accidentes_mortales_año_anterior) * 100

    # Agregar los resultados al DataFrame
    kpi3_comuna1_df = pd.concat([kpi3_comuna1_df, pd.DataFrame({'Año': [año_actual], 'Accidentes_Mortales': [accidentes_mortales_año_actual], 'Cambio': [cambio], 'Cambio_Porcentual': [cambio_porcentual]})], ignore_index=True)


(kpi3_comuna1_df)

```

La comuna 1, si bien tiene un pequeño aumento en 2018 como la tendencia general de todas las comunas, logra reducir sus victimas en los demas años, sorprendentementes tambien en
el año 2021 donde hubo un aumento general de victimas en CABA.
